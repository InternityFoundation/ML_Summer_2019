```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
dataset=pd.read_csv('countries.csv')
dataset
```

```python
#compare the population growth in the Albania and Algeria
```

```python
Algeria = dataset[dataset.country=='Algeria']
Algeria
```

```python
Albania = dataset[dataset.country=='Albania']
Albania
```

```python
plt.plot(Algeria.year , Algeria.population)
plt.plot(Albania.year , Albania.population)
plt.legend(['Algeria','Albania'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```
